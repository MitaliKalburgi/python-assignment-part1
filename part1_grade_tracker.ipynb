**Task 1 - Data Parsing & Profile Cleaning**

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
# Loop through raw_students and for each student, produce a cleaned version

Name = ""
for student in raw_students:
  name = student["name"].strip().title()                        # Step 1: Clean the name
  roll = int(student["roll"])                                   # Step 2: Convert roll to integer
  marks = [int(x) for x in student["marks_str"].split(",")]     # Step 3: Convert marks string to list
  valid = all(word.isalpha() for word in name.split())          # Step 4: Validate name (only alphabets allowed)
  print("======================================")               # Step 5: Print a formatted profile card for each cleaned student using f-strings
  if valid:
    print(f"Student : {name}   ✓ Valid name")
  else:
    print(f"Student : {name}   ✗ Invalid name")
  print(f"Roll No : {roll}")
  print(f"Marks   : {marks}")
  print("======================================")

  # print the name in ALL CAPS and lowercase for the student with roll number 103.

  if roll == 103:
    Name = name
print(Name.upper())
print(Name.lower())
```

**Task 2 - Marks Analysis Using Loops & Conditionals**

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

```python
print("----------------------------")
print(f"{'Subject':<5}   {'Marks':<5}    {'Grade'}")
print("----------------------------")
for i in range(len(subjects)):                       # Using a for loop
  subject = subjects[i]
  mark = marks[i]
  if mark >= 90 and mark <=100:                      # Using conditional statements
    grade = "A+"
  elif mark >= 80 and mark <= 89:
    grade = "A"
  elif mark >= 70 and mark <= 79:
    grade = "B"
  elif mark >= 60 and mark <= 69:
    grade = "C"
  else:
    grade = "F"
  print(f"{subject:<12}{mark:<10}{grade}")           # Printing each subject alongside its marks and a grade label in a table format
print("----------------------------")
```

```python
# Calculating Total marks & Average marks
Total_Marks = sum(marks)
Average_Marks = round(sum(marks)/len(marks) , 2)    # rounded to 2 places

# Finding Highest scoring subject & Lowest scoring subject
max_mark = max(marks)                              # using max to find the larget no.
max_index = marks.index(max_mark)                  # using index to find the index of the largest no.
max_subject = subjects[max_index]
min_mark = min(marks)                              # using min to find the smallest no.
min_index = marks.index(min_mark)                  # using index to find the index of the smallest no.
min_subject = subjects[min_index]

# Printing Total_Marks, Average_Marks, Highest_scoring_subject, Lowest_scoring_subject
print("Calculations:-")
print(f"Total Marks              : {Total_Marks}")
print(f"Average Marks            : {Average_Marks}")
print(f"Highest Scoring Subject  : {max_subject}, {max_mark}")
print(f"Lowest Scoring Subject   : {min_subject}, {min_mark}")
```

```python
# Using a While loop for Marks Entry System
count_new = 0                        # to count how many new subjects are added
while True:

    # Ask for subject name
    subject = input("\nEnter subject name (or type 'done' to stop): ")

    # Stop condition
    if subject.lower() == "done":
        break

    # Check Duplicates
    if subject.lower() in [s.lower() for s in subjects]:
        print("Subject already exists! Enter a different subject.")
        continue

    # Check if subject name has only letters
    if not subject.isalpha():
        print("Subject name must contain only letters.")
        continue

    # Ask for marks
    mark_input = input("Enter marks (0–100): ")

    # Check if input is numeric
    if not mark_input.isdigit():
        print("Invalid input! Please enter numeric marks.")
        continue

    mark = int(mark_input)

    # Check if marks are within valid range
    if mark < 0 or mark > 100:
        print("Marks must be between 0 and 100.")
        continue   # skip invalid entry

    # If valid → add to list
    subjects.append(subject)
    marks.append(mark)
    count_new += 1   # increase counter
# Calculate updated average
new_total = sum(marks)
new_avg = round(new_total / len(marks), 2)

print("\nNew subjects added:", count_new)
print("Updated average:", new_avg)
```

**Task 3 - Class Performance Summary**

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

```python
results = []                                     # to store final data
# Loop through class_data and for each student
for name, marks in class_data:

  # Calculate average
  average = round(sum(marks)/len(marks), 2)

  # Assign status
  if average >= 60:
    status = "Pass"
  else:
    status = "Fail"

  #Store result
  results.append((name, average, status))
```

```python
# Print the formatted class report
print("Name              |Average|Status")
print("------------------------------------")
for name, average, status in results:
  print(f"{name:<18}|{average:<7}|{status}")
```

```python
# Summary
pass_count = 0
fail_count = 0
total_avg = 0
topper_name = ""
topper_avg = 0
for name, average, status in results:
  # Count pass or fail
  if status == "Pass":
    pass_count += 1
  else:
    fail_count += 1

  # Add to total for class average
  total_avg += average

  # Find topper
  if average > topper_avg:
    topper_avg = average
    topper_name = name

# Find class average
class_avg = round(total_avg / len(results), 2)

# Print Summary
print("\nClass Summary:")
print(f"Passed Students : {pass_count}")
print(f"Failed Students : {fail_count}")
print(f"Class Topper    : {topper_name} with average {topper_avg}")
print(f"Class Average   : {class_avg}")
```

**Task 4 - String Manipulation Utility**

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning. "
```

```python
# Step 1: Strip leading and trailing whitespace from essay
clean_essay = essay.strip()

# Step 2: Title Case
print(clean_essay.title())

# Step 3: Count the word 'python' in clean_essay
print(clean_essay.count("python"))

# Step 4: Replace 'python'
print(clean_essay.replace("python", "Python 🐍"))

# Step 5: Split into sentences
sentences = clean_essay.split(". ")
print(sentences)
# Step 6
# Numbered Sentences
for i in range(len(sentences)):
  sentence = sentences[i]

# Ensure sentence ends with '.'
  if not sentence.endswith('.'):
    sentence += '.'
  print(f"{i+1}. {sentence}")
```
